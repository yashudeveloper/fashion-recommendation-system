```python
filtered_stylesdf['category'] = filtered_stylesdf['gender'] + ' ' + filtered_stylesdf['articleType']
```

```python
filtered_stylesdf.drop(columns=['gender', 'articleType'], inplace=True)
```

```python
filtered_stylesdf['product_title'] = (
    filtered_stylesdf['baseColour'].astype(str) + ' ' +
    filtered_stylesdf['season'].astype(str) + ' ' +
    filtered_stylesdf['year'].astype(str) + ' ' +
    filtered_stylesdf['usage'].astype(str) + ' ' +
    filtered_stylesdf['productDisplayName'].astype(str)
)
```

```python
filtered_stylesdf.drop(columns=['baseColour', 'season', 'year', 'usage'], inplace=True)
```

```python
filtered_stylesdf
```

```python
# http://assets.myntassets.com/v1/images/style/properties/45ddbc6a15140556214e15923244755b_images.jpg everything after property
filtered_stylesdf.drop(columns=['id','masterCategory', 'subCategory'], inplace=True)
```

```python
# http://assets.myntassets.com/v1/images/style/properties/45ddbc6a15140556214e15923244755b_images.jpg everything after property
filtered_stylesdf['image_path'] = filtered_stylesdf['link'].str.split('properties/').str[1]
```
