```python
from PIL import Image

instruction = '''You are a fashion expert. Suggest complementary outfit items for the item shown in the image. Use short keywords. Start with "Suggestions:" and format as:
- Item: <Name>
- Colors: <Color Suggestions>
- Suggestions:
  - <Item 1> - <Color>
  - <Item 2> - <Color>
'''

def convert_to_conversation(sample):
    if not isinstance(sample, dict):
        raise TypeError(f"Expected 'sample' to be a dictionary, but got {type(sample).__name__}")

    image_path = sample.get("image_path", "").lstrip("/")
    recommendation = sample.get("recommendation", "No recommendation available.")

    try:

        image = Image.open(image_path).convert("RGB")
    except Exception as e:
        print(f"Warning: Could not load image at '{image_path}'. Error: {e}")
        image = None  # Optionally, provide a default image

    conversation = [
        {
            "role": "user",
            "content": [
                {"type": "text", "text": instruction},
                {"type": "image", "image": image} if image else {"type": "text", "text": "Image not available."}
            ]
        },
        {
            "role": "assistant",
            "content": [
                {"type": "text", "text": recommendation}
            ]
        }
    ]

    return {"messages": conversation}
```

```python
from fastapi import FastAPI, HTTPException
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel
import torch
from transformers import AutoTokenizer, TextStreamer
from PIL import Image
import requests
from io import BytesIO
import re
from pyngrok import ngrok
import uvicorn
import nest_asyncio
nest_asyncio.apply()
# Initialize FastAPI app
app = FastAPI()

# Apply CORS middleware
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

# Assume base model is already loaded as `model` and `tokenizer`

# Define request model
class ImageRequest(BaseModel):
    image_url: str

@app.post("/generate-outfit")
def generate_outfit(request: ImageRequest):
    try:
        # Fetch image from URL
        response = requests.get(request.image_url)
        if response.status_code != 200:
            raise HTTPException(status_code=400, detail="Invalid image URL")

        image = Image.open(BytesIO(response.content)).convert("RGB")

        # Define prompt
        instruction = '''You are a fashion expert. Suggest complementary outfit items for the item shown in the image. Use short keywords. Start with "Suggestions:"
        '''

        # Prepare input
        messages = [{"role": "user", "content": [{"type": "image"}, {"type": "text", "text": instruction}]}]
        input_text = tokenizer.apply_chat_template(messages, add_generation_prompt=True)

        # Tokenize input
        inputs = tokenizer(image, input_text, add_special_tokens=False, return_tensors="pt").to("cuda")

        # Generate response
        text_streamer = TextStreamer(tokenizer, skip_prompt=True)
        output = model.generate(**inputs, streamer=text_streamer, max_new_tokens=128, use_cache=True, temperature=1.5, min_p=0.1)

        # Decode and format response
        decoded_output = tokenizer.decode(output[0], skip_special_tokens=True)

        # Extract text after 'Suggestions:'
        print("yha pr :",decoded_output)
        match = re.search(r'^(?:.*?\n){3}(.*)', decoded_output, re.DOTALL)
        filtered_output = match.group(1).strip() if match else decoded_output
        print("yha pr :",filtered_output)
        return {"suggestions": filtered_output}

    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

# Start Ngrok tunnel and run the API
port = 4000
ngrok_tunnel = ngrok.connect(port)
print(f"Public URL: {ngrok_tunnel.public_url}")

uvicorn.run(app, host="0.0.0.0", port=port)
```
